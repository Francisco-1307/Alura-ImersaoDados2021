```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
sns.set()

# Dados experimentos
dados_exp = pd.read_csv("./dados_experimentos.csv")

# Renomear colunas (tirando o hífen -)
colunas = list(dados_exp.columns)
novas_colunas = [col.replace('-','') for col in colunas]
novas_colunas
dados_exp.columns = novas_colunas

# Alterando o nome da coluna 'droga'
dados_exp.rename(columns = {'droga':'composto'}, inplace=True)

# Dados resultados
dados_resultados = pd.read_csv("./dados_resultados.csv")

# Adicionando colunas extras
dados_resultados['n_moa'] = dados_resultados.sum(axis = 1)
dados_resultados['ativo_moa'] = (dados_resultados['n_moa'] != 0)

# Merge dos datasets
dados_ExpRes = pd.merge(dados_resultados[['id','n_moa','ativo_moa']], dados_exp, on='id')

# Coluna nova 'eh_controle'
dados_ExpRes['eh_controle'] = (dados_ExpRes['tratamento'] == 'com_controle')

# Colunas novas 'duracao_xh'
for duracao in dados_ExpRes['tempo'].unique():
    coluna_nova = 'duracao_{}h'.format(duracao)
    dados_ExpRes[coluna_nova] = dados_ExpRes['tempo'] == duracao
    
colunas_nova_ordem = dados_ExpRes.columns.to_list()
popped_list = [colunas_nova_ordem.pop() for popped in range(4)]
for popped in popped_list:
    colunas_nova_ordem.insert(colunas_nova_ordem.index('g0'),popped)
    
dados_ExpRes = dados_ExpRes[colunas_nova_ordem]
```

```python
dados_ExpRes
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
# Intanciando o modelo
model_rlogistica = LogisticRegression(max_iter = 1000)

x = dados_ExpRes.select_dtypes('float64')
y = dados_ExpRes['ativo_moa']#.astype('int')
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size = 0.2, stratify = y, random_state = 726)

model_rlogistica.fit(x_treino, y_treino)
model_rlogistica.score(x_teste, y_teste)
```

```python
from sklearn.dummy import DummyClassifier
from sklearn.metrics import accuracy_score
```

```python
model_dummy = DummyClassifier(strategy='most_frequent')
model_dummy.fit(x_treino, y_treino)
y_previsao = model_dummy.predict(x_teste)
accuracy_score(y_teste, y_previsao)
```

```python
dados_ExpRes['ativo_moa'].value_counts(normalize = True)
```

```python
from sklearn.tree import DecisionTreeClassifier

x = dados_ExpRes.select_dtypes('float64')
y = dados_ExpRes['ativo_moa']#.astype('int')
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size = 0.2, stratify = y, random_state = 726)

modelo_arvore = DecisionTreeClassifier(max_depth = 3)
modelo_arvore.fit(x_treino, y_treino)
modelo_arvore.score(x_teste, y_teste)
```

```python
from sklearn import tree

fig, ax = plt.subplots(figsize = (15,10), facecolor = 'k')
tree.plot_tree(modelo_arvore,
               ax = ax,
               fontsize = 10,
               rounded = True,
               filled = True,
               feature_names = x_treino.columns,
               class_names = ["Não ativado", "Ativado"])
plt.show()
```

```python
x = dados_ExpRes.select_dtypes('float64')
y = dados_ExpRes['ativo_moa']#.astype('int')
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size = 0.2, stratify = y, random_state = 726)

teste  = []
treino = []
for i in range(1,15):
    modelo_arvore = DecisionTreeClassifier(max_depth = i)
    modelo_arvore.fit(x_treino, y_treino)
    treino.append(modelo_arvore.score(x_treino, y_treino))
    teste.append(modelo_arvore.score(x_teste, y_teste))
```

```python
# Overfitting
sns.lineplot(x = range(1,15), y = teste, label = 'teste')
sns.lineplot(x = range(1,15), y = treino, label = 'treino')
```
