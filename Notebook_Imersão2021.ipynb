```python
import pandas as pd

dados = pd.read_csv("./dados_experimentos.csv")

# Renomear colunas (tirando o hífen -)
colunas = list(dados.columns)
novas_colunas = [col.replace('-','') for col in colunas]
novas_colunas
dados.columns = novas_colunas

# Alterando o nome da coluna 'droga'
dados.rename(columns = {'droga':'composto'}, inplace=True)
```

```python
dados
```

```python
# Identificando os 5 compostos com mais experimentos
top5_compostos = dados['composto'].value_counts()[0:5]
top5_compostos 
```

```python
# Fazendo uma seleção no data frame
top5_index = top5_compostos.index
top5_compostosDF = dados.query('composto in @top5_index')
top5_compostosDF
```

```python
# importando bibliotecas para plots
import seaborn as sns
import matplotlib.pyplot as plt 
sns.set() # configuração padrão do sns
```

```python
# Plotando o top 5 compostos
plt.figure(figsize = (8,6))
ax = sns.countplot(x = 'composto',
                  data = top5_compostosDF,
                  order = top5_index,
                  palette = 'Blues_r') # Desafio 01
ax.set_title("Top 5 Compostos")
ax.set_ylabel("Contagem")
ax.set_xlabel("Código do composto")
plt.show()
```

```python
import pandas as pd

dados = pd.read_csv("./dados_experimentos.csv")

# Renomear colunas (tirando o hífen -)
colunas = list(dados.columns)
novas_colunas = [col.replace('-','') for col in colunas]
novas_colunas
dados.columns = novas_colunas

# Alterando o nome da coluna 'droga'
dados.rename(columns = {'droga':'composto'}, inplace=True)
```

```python
dados
```

```python
# Utilizando o crosstab para cruzar dados de colunas diferentes com base na contagem de experimentos.
pd.crosstab(index = dados['dose'], columns = dados['tempo'])
```

```python
pd.crosstab(index = [dados['dose'],dados['tempo']], columns = dados['tratamento'])
```

```python
pd.crosstab(index = [dados['dose'],dados['tempo']], columns = dados['tratamento'], normalize = 'index')
```

```python
# Podemos usar outras funções para agregar os dados. 
pd.crosstab(index = [dados['dose'],dados['tempo']], columns = dados['tratamento'],
            values = dados['g0'], aggfunc = 'mean') # usando a média a expressão de g0
```

```python
pd.crosstab(index = [top5_compostosDF['dose'],top5_compostosDF['tempo']], columns = [top5_compostosDF['composto']])
```

```python
# O composto cacb2b860 é o usado para o controle!
pd.crosstab(index = [dados['tratamento']], columns = [top5_compostosDF['composto']])
```

```python
dados.loc[:,'g0':'g771'].corr()
```

```python
from string import ascii_letters
import numpy as np

corr = dados.loc[:,'g0':'g50'].corr()

# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

```python
# Correlação entre os valores do c's, que quantifica a viabilidade celular, isso é, taxa de sobrevivência.
corr_celular = dados.loc[:,'c0':'c50'].corr()

# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr_celular, dtype=bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr_celular, mask=mask, cmap=cmap, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

**Desafio 05**: Calcular e analisar a correlação entre os g's e c's

```python
dados_g = dados.loc[:,'g0':'g50']
dados_c = dados.loc[:,'c0':'c99']

dados_gc = pd.concat([dados_g, dados_c], axis = 1)

corr_gc = dados_gc.corr().loc['c0':'c99','g0':'g50']
corr_gc
```

```python
# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(15, 30))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio

sns.heatmap(corr_gc, cmap=cmap, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```
